# External Model (p)

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt

start_time = 0
end_time = 5
ts = 0.01

t = np.arange(start_time,end_time,ts)


def dN_dt(N, NA, p, q):
    return (p + q * (N/NA)) * (NA - N)

NA = 1

# plotting the spread or diffusion of innovation
derivative = []
p_values = [10,7,5,1,0.75,0.5,0.25]
q = 0
for p in p_values:
  N = np.zeros_like(t)
  N[0] = 0.1
  der = []
  der.append(0)

  for i in range(1, len(t)):
    N[i] = N[i-1] + (ts*dN_dt(N[i-1], NA, p, q))
    der.append(dN_dt(N[i-1], NA, p, q))
  derivative.append(der)
  string = 'p = ' + str(p)
  
  plt.plot(t,N,label=string)
  plt.ylabel('Market Share Percent(in fraction)')
  plt.xlabel('time')

plt.title('External Infuence Model (p)')
plt.legend()
plt.show()

# plotting derivative for bell shaped curve
for i in range(len(p_values)):
  string = 'p = ' + str(p_values[i])
  plt.plot(t,derivative[i],label = string)

plt.xlabel('time')
plt.ylabel('dN/dt')
plt.title('External Influence Model (p))')
plt.legend()
plt.show()
```

## Bass Model

```python
start_time = 0
end_time = 60
ts = 0.01

t = np.arange(start_time,end_time,ts)


def dN_dt(N, NA, p, q):
    return (p + q * (N/NA)) * (NA - N)

NA = 1

# plotting the spread or diffusion of innovation
derivative = []
q_values = [0.21566, 0.17110, 0.25105, 0.29695, 0.41861, 0.35688, 0.33790, 0.24387, 0.30145, 0.32791, 0.65410]
p_values = [0.0026167, 0.018119, 0.027877, 0.017703, 0.010399, 0.017206, 0.0091837, 0.005876, 0.017135, 0.028632, 0.024796]
for j in range(len(q_values)):
  N = np.zeros_like(t)
  N[0] = 0.0
  der = []
  der.append(0)

  for i in range(1, len(t)):
    N[i] = N[i-1] + (ts*dN_dt(N[i-1], NA, p_values[j], q_values[j]))
    der.append(dN_dt(N[i-1], NA, p_values[j], q_values[j]))
  derivative.append(der)
  string = 'p = ' + str(p_values[j]) + ', q = ' + str(q_values[j])
  
  plt.figure(j)
  plt.plot(t,N,label='fraction of users')
  plt.plot(t,der,label = 'market penetration rate')
  plt.ylabel('Market Share Percent(in fraction)')
  plt.xlabel('time')
  plt.title('Bass Model '+string)
  plt.legend()
  plt.show()
```

## Modified bass model

```python
start_time = 0
end_time = 30
ts = 0.01

t = np.arange(start_time,end_time,ts)


def dN_dt(N, NA, p, q, beta):
    return (p + q * ((N/NA) ** (beta+1)) ) * (NA - N)

NA = 1

beta = -0.25

# plotting the spread or diffusion of innovation
derivative = []
q_values = [0.5,0.5,0.05,0.0005]
p_values = [0.0005,0.05,0.5,0.5]
for j in range(len(q_values)):
  N = np.zeros_like(t)
  N[0] = 0.001
  der = []
  der.append(0)

  for i in range(1, len(t)):
    N[i] = N[i-1] + (ts*dN_dt(N[i-1], NA, p_values[j], q_values[j], beta))
    der.append(dN_dt(N[i-1], NA, p_values[j], q_values[j], beta))
  derivative.append(der)
  string = 'p = ' + str(p_values[j]) + ', q = ' + str(q_values[j])
  
  plt.figure(j)
  plt.plot(t,N,label='fraction of users')
  plt.plot(t,der,label = 'market penetration rate')
  plt.ylabel('Market Share Percent(in fraction)')
  plt.xlabel('time')
  plt.title('Modified Bass Model '+string)
  plt.legend()
  plt.show()
```
