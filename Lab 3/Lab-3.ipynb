```python
import numpy as np
import math
import matplotlib.pyplot as plt
t_final = 2000
t_step = 0.1
k=1000
r=0.5
h=63.75
n = int(t_final/t_step)
x = np.zeros(n)
y = np.zeros(n)
x0 = 300  # Initial population 
x[0] = x0
y[0] = x0
```

# Dimensionless Equations

```python
h = (r*k/4) 
h_tild = 0.5
A_tilde = 0.2
N = np.zeros(n)
N[0] = x0/k
for i in range(1,int(n)):
    N[i] = N[i-1] + ((N[i-1]*(1-N[i-1])*(N[i-1]/A_tilde - 1))- h_tild*N[i-1])*t_step*r     
```

```python
fig, ax = plt.subplots(figsize=[8,8])
t = np.linspace(0,t_final*r, int(n))
plt.plot(r*t, N, label='normalized population - n(t)',ls='-',lw=1.2, color='k')
#plt.plot(t, N_2, label='Population (N) - h = {}'.format(h2),ls='--',lw=1.2, color='k')
plt.xlabel('\u03C4', fontsize=15, color='k')
plt.ylabel('Normalized population (n)', fontsize=15, color='k')
plt.title(r'Population growth: k = {}, $\~A$ = {}, $\~\epsilon$ = {} , n(0) = {}'.format(k,A_tilde, h_tild, N[0]), fontsize=15, color='k')
#plt.title(r'Population growth with threshold: k = {}, $\~A$ = {}, $\~h$ = {}, n(0) = {}'.format(k,A_tilde,h_tild, N[0]), fontsize=15, color='k')
#plt.axhline(k,label='Carrying capacity', color='k', ls="--", lw=1)=
#plt.axhline(k/2,label='Carrying capacity/2', color='k', ls='-', lw=1)
plt.legend(fontsize=15, loc='best')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid( color='k', alpha=0.5, ls=':')
plt.show()
```
